```python
import pandas as pd
import json
```

```python
with open('/content/t20_wc_match_results.json') as f:
  data = json.load(f)
data[:5]
```

```python
data[0]['matchSummary']
```

```python
dataFrame = pd.DataFrame(data[0]['matchSummary'])
dataFrame.head()
```

```python
dataFrame.rename({'scorecard': 'match_id'}, axis=1,inplace=True)
```

```python
dataFrame.head()
```

```python
match_ids_dict = {}
for index, row in dataFrame.iterrows():
  key1 = row['team1'] + ' Vs ' + row['team2']
  key2 = row['team2'] + ' Vs ' + row['team1']

  match_ids_dict[key1] = row['match_id']
  match_ids_dict[key2] = row['match_id']

match_ids_dict
```

### Batting Summary

```python
with open('/content/t20_wc_batting_summary.json') as b:
  data_bat = json.load(b)
data_bat[:5]
```

```python
all_records = []
for rec in data_bat:
  all_records.extend(rec['battingSummary'])

dataFrameBat= pd.DataFrame(all_records)
dataFrameBat.head()
```

```python
dataFrameBat.tail()
```

```python
dataFrameBat["out/not_out"] = dataFrameBat.dismissal.apply(lambda x: "out" if len(x) > 0 else "not_out")
dataFrameBat.head(10)
```

```python
dataFrameBat = dataFrameBat.drop('dismissal', axis=1)
dataFrameBat.head(10)
```

```python
dataFrameBat['batsmanName'] = dataFrameBat['batsmanName'].apply(lambda x: x.replace('â€', ''))
dataFrameBat['batsmanName'] = dataFrameBat['batsmanName'].apply(lambda x: x.replace('\xa0', ''))
dataFrameBat['batsmanName'] = dataFrameBat['batsmanName'].apply(lambda x: x.replace('†', ''))
dataFrameBat.head(10)
```

```python
match_ids_dict['Bangladesh Vs India']
```

```python
dataFrameBat['match_id'] = dataFrameBat['match'].map(match_ids_dict)

dataFrameBat.head()
```

```python
# dataFrameBat = dataFrameBat.drop('mach_id',axis=1)
```

```python
dataFrameBat.head()
```

```python
dataFrameBat.to_csv('/content/summary/temp.csv', index=False)
```

### Bowling Summary

```python
with open('/content/t20_wc_bowling_summary.json') as bl:
  data_ball = json.load(bl)
data_ball[:2]
```

```python
all_ball_records = []
for rec in data_ball:
  all_ball_records.extend(rec['bowlingSummary'])

all_ball_records[:2]
```

```python
dataFrameBall = pd.DataFrame(all_ball_records)
print(dataFrameBall.shape)
dataFrameBall.head()
```

```python
dataFrameBall['match_id'] = dataFrameBall['match'].map(match_ids_dict)
dataFrameBall.head()
```

```python
dataFrameBall.to_csv('/content/summary/tempBall.csv', index=False)
```
